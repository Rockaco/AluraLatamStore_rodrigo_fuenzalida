#1. Análisis de facturación



```python
# Calcula el ingreso total para cada tienda sumando la columna 'Precio'
# Se accede a la columna 'Precio' de cada DataFrame (tienda, tienda2, tienda3, tienda4)
# y se utiliza el método .sum() para obtener la suma de los valores en esa columna
ingreso_total_tienda1 = tienda['Precio'].sum()
ingreso_total_tienda2 = tienda2['Precio'].sum()
ingreso_total_tienda3 = tienda3['Precio'].sum()
ingreso_total_tienda4 = tienda4['Precio'].sum()

# Imprime el ingreso total calculado para cada tienda
# Se utilizan f-strings para formatear la salida y mostrar los resultados de manera clara
print(f"Ingreso total Tienda 1: {ingreso_total_tienda1:,.2f}") # Muestra el ingreso de la tienda 1 formateado con comas y dos decimales
print(f"Ingreso total Tienda 2: {ingreso_total_tienda2:,.2f}") # Muestra el ingreso de la tienda 2 formateado
print(f"Ingreso total Tienda 3: {ingreso_total_tienda3:,.2f}") # Muestra el ingreso de la tienda 3 formateado
print(f"Ingreso total Tienda 4: {ingreso_total_tienda4:,.2f}") # Muestra el ingreso de la tienda 4 formateado
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Crear un DataFrame con los ingresos totales por tienda
# Se crea un diccionario donde las claves son los nombres de las tiendas y los valores son los ingresos totales calculados previamente
ingresos_por_tienda = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'], # Nombres de las tiendas
    'Ingreso Total': [ingreso_total_tienda1, ingreso_total_tienda2, ingreso_total_tienda3, ingreso_total_tienda4] # Ingresos totales calculados
})

# Ordenar los datos por ingreso total para una mejor visualización
ingresos_por_tienda = ingresos_por_tienda.sort_values('Ingreso Total', ascending=False) # Ordena el DataFrame en orden descendente por 'Ingreso Total'

# Configurar el estilo de los gráficos (estilo "Trash Metal")
plt.style.use('dark_background') # Usa un fondo oscuro
plt.rcParams['text.color'] = 'cyan' # Color del texto cian
plt.rcParams['axes.labelcolor'] = 'cyan' # Color de las etiquetas de los ejes cian
plt.rcParams['xtick.color'] = 'cyan' # Color de los ticks del eje x cian
plt.rcParams['ytick.color'] = 'cyan' # Color de los ticks del eje y cian
plt.rcParams['axes.edgecolor'] = 'cyan' # Color de los bordes de los ejes cian


# Crear el gráfico de barras horizontales
plt.figure(figsize=(10, 6)) # Define el tamaño de la figura
plt.barh(ingresos_por_tienda['Tienda'], ingresos_por_tienda['Ingreso Total'], color=['cyan', 'magenta', 'yellow', 'lime']) # Crea el gráfico de barras horizontales con colores llamativos

# Añadir etiquetas y título
plt.xlabel('Ingreso Total') # Etiqueta del eje x
plt.ylabel('Tienda') # Etiqueta del eje y
plt.title('Ingreso Total por Tienda') # Título del gráfico

# Formatear los valores del eje x para que sean más legibles (con comas)
plt.gca().xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: '{:,.0f}'.format(x))) # Formatea los números con comas y sin decimales

# Mostrar el gráfico
plt.show() # Muestra la figura
```

# 2. Ventas por categoría

```python
# Calcula la cantidad de productos vendidos por categoría para la Tienda 1
# Agrupa el DataFrame 'tienda' por la columna 'Categoría del Producto'
# Utiliza .size() para contar el número de filas en cada grupo (cada categoría)
ventas_por_categoria_tienda1 = tienda.groupby('Categoría del Producto').size().sort_values(ascending=False) # Ordena los resultados de forma descendente

# Muestra los resultados para la Tienda 1
print("--- Ventas por Categoría - Tienda 1 ---") # Encabezado informativo
display(ventas_por_categoria_tienda1) # Muestra la serie con las ventas por categoría

print("\n" + "="*40 + "\n") # Separador

# Calcula la cantidad de productos vendidos por categoría para la Tienda 2
# Agrupa el DataFrame 'tienda2' por la columna 'Categoría del Producto'
ventas_por_categoria_tienda2 = tienda2.groupby('Categoría del Producto').size().sort_values(ascending=False) # Ordena los resultados

# Muestra los resultados para la Tienda 2
print("--- Ventas por Categoría - Tienda 2 ---") # Encabezado informativo
display(ventas_por_categoria_tienda2) # Muestra los resultados

print("\n" + "="*40 + "\n") # Separador

# Calcula la cantidad de productos vendidos por categoría para la Tienda 3
# Agrupa el DataFrame 'tienda3' por la columna 'Categoría del Producto'
ventas_por_categoria_tienda3 = tienda3.groupby('Categoría del Producto').size().sort_values(ascending=False) # Ordena los resultados

# Muestra los resultados para la Tienda 3
print("--- Ventas por Categoría - Tienda 3 ---") # Encabezado informativo
display(ventas_por_categoria_tienda3) # Muestra los resultados

print("\n" + "="*40 + "\n") # Separador

# Calcula la cantidad de productos vendidos por categoría para la Tienda 4
# Agrupa el DataFrame 'tienda4' por la columna 'Categoría del Producto'
ventas_por_categoria_tienda4 = tienda4.groupby('Categoría del Producto').size().sort_values(ascending=False) # Ordena los resultados

# Muestra los resultados para la Tienda 4
print("--- Ventas por Categoría - Tienda 4 ---") # Encabezado informativo
display(ventas_por_categoria_tienda4) # Muestra los resultados
```

# 5. Envío promedio por tienda

```python
# Calcula el costo de envío promedio para cada tienda
# Se accede a la columna 'Costo de envío' de cada DataFrame (tienda, tienda2, tienda3, tienda4)
# y se utiliza el método .mean() para obtener el promedio de los costos de envío
costo_envio_promedio_tienda1 = tienda['Costo de envío'].mean()
costo_envio_promedio_tienda2 = tienda2['Costo de envío'].mean()
costo_envio_promedio_tienda3 = tienda3['Costo de envío'].mean()
costo_envio_promedio_tienda4 = tienda4['Costo de envío'].mean()

# Imprime el costo de envío promedio calculado para cada tienda
# Se utilizan f-strings para formatear la salida y mostrar los resultados con dos decimales
print(f"Costo de envío promedio Tienda 1: {costo_envio_promedio_tienda1:,.2f}") # Muestra el costo de envío promedio de la tienda 1 formateado
print(f"Costo de envío promedio Tienda 2: {costo_envio_promedio_tienda2:,.2f}") # Muestra el costo de envío promedio de la tienda 2 formateado
print(f"Costo de envío promedio Tienda 3: {costo_envio_promedio_tienda3:,.2f}") # Muestra el costo de envío promedio de la tienda 3 formateado
print(f"Costo de envío promedio Tienda 4: {costo_envio_promedio_tienda4:,.2f}") # Muestra el costo de envío promedio de la tienda 4 formateado
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Crear un DataFrame con los costos de envío promedio por tienda
# Se crea un diccionario donde las claves son los nombres de las tiendas y los valores son los costos de envío promedio calculados previamente
costos_envio_promedio = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'], # Nombres de las tiendas
    'Costo de Envío Promedio': [costo_envio_promedio_tienda1, costo_envio_promedio_tienda2, costo_envio_promedio_tienda3, costo_envio_promedio_tienda4] # Costos de envío promedio calculados
})

# Ordenar los datos por costo de envío promedio para una mejor visualización
costos_envio_promedio = costos_envio_promedio.sort_values('Costo de Envío Promedio', ascending=False) # Ordena el DataFrame en orden descendente por 'Costo de Envío Promedio'

# Configurar el estilo de los gráficos (estilo "Trash Metal")
plt.style.use('dark_background') # Usa un fondo oscuro
plt.rcParams['text.color'] = 'cyan' # Color del texto cian
plt.rcParams['axes.labelcolor'] = 'cyan' # Color de las etiquetas de los ejes cian
plt.rcParams['xtick.color'] = 'cyan' # Color de los ticks del eje x cian
plt.rcParams['ytick.color'] = 'cyan' # Color de los ticks del eje y cian
plt.rcParams['axes.edgecolor'] = 'cyan' # Color de los bordes de los ejes cian

# Crear el gráfico de barras
plt.figure(figsize=(10, 6)) # Define el tamaño de la figura
plt.bar(costos_envio_promedio['Tienda'], costos_envio_promedio['Costo de Envío Promedio'], color=['cyan', 'magenta', 'yellow', 'lime']) # Crea el gráfico de barras con colores llamativos

# Añadir etiquetas y título
plt.xlabel('Tienda') # Etiqueta del eje x
plt.ylabel('Costo de Envío Promedio') # Etiqueta del eje y
plt.title('Costo de Envío Promedio por Tienda') # Título del gráfico

# Formatear los valores del eje y para que sean más legibles (con comas)
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda y, _: '{:,.0f}'.format(y))) # Formatea los números con comas y sin decimales

# Mostrar el gráfico
plt.show() # Muestra la figura
```
